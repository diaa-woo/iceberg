### 주의점

예를 다른 변수에 심을려면 그냥 들고와서는 안됨. ndarray 자료형에 내장되어 있는 기능 중 `copy`를 사용해야 함. 단순히 b = a와 같이 대입하면 a의 내용이 저장된 곳의 참조 주소가 전달됨.(포인터 개념, 즉 ndarray는 주소값들이 뭉쳐져있고 값을 가져오는 건 포인터 참조라고 생각하면 편할 듯) 즉, a와 b가 연결되어 둘 중 하나라도 변경시 같이 변경됨.

```python
a = np.array([1, 1])
b = a  # 주소 복사
print('a = ' + str(a))
print('b = ' + str(b))
b[0] = 100
print('a = ' + str(a))
print('b = ' + str(b))
```

```python
a = np.array([1, 1])
b = a.copy()  # 독립 메모리 공간 할당
print('a = ' + str(a))
print('b = ' + str(b))
b[0] = 100
print('a = ' + str(a))
print('b = ' + str(b))
```

인수랑 반환값 다 줄 수 있다는거 알제?

```python
def my_func3(D):
  m = np.mean(D)
  s = np.std(D)
  return m, s
```

```python
data = np.random.randn(100)
data_mean, data_std = my_func3(data)
print('mean: {0:3.2f}, std: {1:3.2f}'.format(data_mean, data_std))
```

반환값이 여러 개여도 하나의 변수 받는거 ㅆㄱㄴ. 글고 받게 되면 그 자료형은 tuple로 받아야함

```python
output = my_func3(data)
print(output)
print(type(output))
print('mean: {0:3.2f}, std: {1:3.2f}'.format(output[0], output[1]))
```

## 파일 저장

사라지면... 가슴 아프지나...

### 하나의 ndarray 형을 저장

`np.save('파일명.npy', 변수명)`을 사용. 파일의 확장자는 **.npy**임. 데이터를 로드하려면 `np.load('파일명.npy')`을 사용

```python
data = np.random.randn(5)
print(data)
np.save('datafile.npy', data)    # 저장
data = []                        # 데이터 삭제
print(data)
data = np.load('datafile.npy')
print(data)
```

### 여러 ndarray형을 저장

여러 ndarray형을 한번에 저장! `np.savez('파일명.npz', 변수명1 = 변수명1, 변수명2 = 변수명2, ...)`

```python
data1 = np.array([1, 2, 3])
data2 = np.array([10, 20, 30])
np.savez('datafile2.npz', data1=data1, data2=data2)
data1 = []                        # 데이터 삭제
data2 = []   
outfile = np.load('datafile2.npz')
print(outfile.files)

data1 = outfile['data1']
data2 = outfile['data2']
print(data1)
print(data2)
```
