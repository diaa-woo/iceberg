# 데이터 프레임의 이해

## 데이터 프레임이란

**데이터 프레임(Data Frame)** 은 데이터를 다룰 때 가장 많이 사용되는 데이터 형태로, **행과 열** 로 구성된 사각형 모양의 표처럼 생겼다. 예를 들어 아래와 같은 데이터들이 존재한다고 가정했을 때,

|성별|연령|학점|연봉|
|---|---|---|---|
|남자|26|3.8|2,700만원|
|여자|42|4.2|4,000만원|
|남자|35|2.6|3,500만원|

이 표는 열 4개와 행 3개로 구성된 데이터 프레임이라고 부른다. 열은 성별, 연령, 학점, 연봉의 4가지 **속성**으로 되어 있으며, 행은 3개의 **정보**를 담고 있다. 한 줄로 표현하면 '4개의 속성에 대한 3명의 자료로 구성된 데이터 프레임' 이라고 할 수 있다.

### '열'은 속성이다
세로로 나연되는 열은 속성을 나타낸다. 열은 **컬럼(Column)** & **변수(Variable)** 라고 부른다.   
만약 어떤 데이터에 100가지 속성의 정보가 들어 있다면, 그 데이터는 100개의 컬럼 또는 100개의 변수를 지니고 있다고 표현할 수 있다.

### '행'은 한 사람의 정보다
가로로 나열되는 행은 각 사람의 정보를 나타낸다. 행은 **로(Row)** & **케이스(Case)** 라고 불린다.   
만약 어떤 데이터에 30명의 정보가 들어 있다면, 그 데이터는 30개의 로 또는 30개의 케이스를 가지고 있다고 표현할 수 있다.

행은 한 사람 및 한 단위의 정보로써, 그 주체는 비단 사람뿐 아니라 도시, 회사 등등 여러 개의 속성을 담은 한 개의 행이 되어줄 수 있다.

### 큰 데이터의 우선순위 : 행 vs 열
데이터가 크다고 표현하는 경우는 행이 많거나, 열이 많은 경우이다. 만일 둘 중 중요한걸 하나를 꼽으라면 어떤 것일까? 바로 열이다.

100명의 데이터를 분석하고 있는데 어느순간 데이터가 10만명으로 늘어난다면 컴퓨터가 느려질 것이다. 이 문제를 해결하는 방법은 당연하게도 컴퓨터의 사양을 높이는 것이다. 메모리 & CPU를 늘려보고 클라우드 서비스 혹은 분산 컴퓨팅 등 컴퓨터 시스템의 사양을 높이는 과정은 하드웨어적 부분이다. 설령 데이터가 늘어난다고 할지 언정 분석 기술이 달라지지는 않을 것이다.

반면 열은 어떨까? 데이터 분석에서는 '학점과 연봉의 관계' & '전공에 따른 연봉의 차이' 등 변수 간의 관계를 다룬다. 그런데 열이 많아지면 변수를 조합할 수 있는 경우의 수가 늘어나버린다. 단순히 몇개의 속성에서는 그저 그러겠지만, 그 수가 기하급수적으로 많아진다면 어떨까. 해당 변수 간의 관계를 알아내는 것이 상당히 복잡해질 것이다. 이렇게 변수가 늘어난다면 단순한 분석 방법으로는 해결하기 힘드므로 여러 변수의 영향을 동시에 고려하는 복잡한 분석 방법을 활용할 필요성이 생긴다.(머신러닝이 주목받는 이유)   
이처럼 변수가 많아지면 적용할 분석 기술이 달라지기 때문에 분석가는 전보다 더 노력을 기울여야 한다. 때문에 데이터의 양, 즉 행보다는 데이터의 다양성, 열을 더 중요시 여긴다.

### 빅데이터?

데이터 분석의 가치는 어떤 현상이 조건에 따라 달라진다는 사실을 발견할 때 생겨난다. '특정 날씨에 어떤 음식이 더 잘 팔린다', '특정 모양의 도로에서 교통사고 발생률이 더 높아진다' 등 이런 조건과 현상의 관계를 잘 밝혀낸다면 해당 현상을 증폭, 또는 억제하는데 큰 도움을 줄 수 있다. 이런 식의 조건과 현상의 관계를 잘 밝혀내려면 데이터가 다양한 변수로 구성되어 있어야 한다. 데이터가 아무리 많더라도 다양한 변수가 담겨 있지 않으면 변수 간의 관련성을 분석할 수 없으므로 의미있는 정보를 찾아낼 수 없다.
